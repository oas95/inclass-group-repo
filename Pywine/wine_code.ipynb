```python
import pandas as pd 
```

```python
# CSV File Name
wine_data = "Data/winemag-data-130k-v2.csv"
```

```python
# Convert to DataFrame
wine_df = pd.read_csv(wine_data, encoding = 'latin-1')
```

```python
wine_df.head()
```

```python
mean = wine_df['points'].groupby(wine_df['country']).mean()

mean_df = pd.DataFrame({"Average Score by Country":mean})
```
